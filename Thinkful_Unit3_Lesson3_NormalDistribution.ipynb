# Unit 3 Lesson 3: Descriptive Statistics and Normality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reference_color="crimson"
line_width = 1
main_color = "mediumaquamarine"
```

```python
# Chi-square distribution with three degrees of freedom
xi = pd.DataFrame()
xi["rand"] = pd.Series(np.random.chisquare(3, 100))

plt.hist(xi["rand"], color=main_color)
plt.axvline(xi["rand"].mean(), linewidth=line_width, color=reference_color)
plt.axvline(xi["rand"].mean() + xi["rand"].std(), color=reference_color, linestyle="dashed")
plt.axvline(xi["rand"].mean() - xi["rand"].std(), color=reference_color, linestyle="dashed")
plt.show()
```

__Notes on the Chi-Square distribution:__ The mean and standard deviations were sort of helpful insofar as most of the area under the curve did indeed appear to fall within one standard deviation of the arithematic mean. However, the mean was not descriptive of a central tendency because this distribution is not symmetrical around the mean the way a normal distribution is. Instead, it is heavily right skewed.

```python
# F distribution
f1 = pd.Series(np.random.f(5, 2, 100))
f2 = pd.Series(np.random.f(100, 2, 100))
f3 = pd.Series(np.random.f(2, 100, 100))
f4 = pd.Series(np.random.f(100, 100, 100))

fbins = 50

fig, axes = plt.subplots(2,2, figsize=(12, 7))

ax1 = axes[0][0]
ax1.hist(f1, color=main_color, bins=fbins)
ax1.axvline(f1.mean(), color=reference_color, linewidth=line_width)
ax1.axvline(f1.mean() + f1.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax1.axvline(f1.mean() - f1.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax1.set_title("ndof = 5, ddof = 2")


ax2 = axes[0][1]
ax2.hist(f2, color=main_color, bins=fbins)
ax2.axvline(f2.mean(), color=reference_color, linewidth=line_width)
ax2.axvline(f2.mean() + f2.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax2.axvline(f2.mean() - f2.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax2.set_title("ndof = 100, ddof = 2")

ax3 = axes[1][0]
ax3.hist(f3, color=main_color, bins=fbins)
ax3.axvline(f3.mean(), color=reference_color, linewidth=line_width)
ax3.axvline(f3.mean() + f3.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax3.axvline(f3.mean() - f3.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax3.set_title("ndof = 2, ddof = 2")

ax4 = axes[1][1]
ax4.hist(f4, color=main_color, bins=fbins)
ax4.axvline(f4.mean(), color=reference_color, linewidth=line_width)
ax4.axvline(f4.mean() + f4.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax4.axvline(f4.mean() - f4.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
ax4.set_title("ndof = 100, ddof = 100")

plt.show()
```

__Notes on F distribution:__ The mean and standard deviations were even less helpful for describing this distribution than the Chi-Square because of how heavily right skewed the distribution is. In part this is because of the degrees of freedom I chose. the larger both degrees of freedom the more normal the distribution becomes and the more helpful these metrics become.

```python
# Binomial distribution
bi = pd.Series(np.random.binomial(15, 0.3, 100))

plt.hist(bi, color=main_color, bins=20)
plt.axvline(bi.mean(), color=reference_color, linewidth=line_width)
plt.axvline(bi.mean() + bi.std(), color=reference_color, linewidth = line_width, linestyle="dashed")
plt.axvline(bi.mean() - bi.std(), color=reference_color, linewidth = line_width, linestyle="dashed")
plt.title("Binomial Distribution, n=15, p=0.3")
plt.show()
```

__Notes on binomial distribution:__ The mean and standard deviation appear to be a better fit for the binomial distribution due to its being roughly symmetrical around its mean. As far as I can tell that's a property of the binomial distribution regardless of the value of n.

```python
# Logistic distribution
logit = pd.Series(np.random.logistic(loc=2, scale=1, size=100))

plt.hist(logit, color=main_color, bins=50)
plt.axvline(logit.mean(), color=reference_color, linewidth=line_width)
plt.axvline(logit.mean() + logit.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.axvline(logit.mean() - logit.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.title("Logistic Distribution, mean=2, scale=1")
plt.show()
```

__Notes on logistic distribution:__ Like the binomial distribution, the mean and standard deviation appear to estimate central tendency and variation from the central tendency relatively well because the distribution is symmetric about its mean.

```python
# Wald distribution
w = pd.Series(np.random.wald(50, 10, 100))

plt.hist(w, color=main_color, bins=20)
plt.axvline(w.mean(), color=reference_color, linewidth=line_width)
plt.axvline(w.mean() + w.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.axvline(w.mean() - w.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.title("Wald Distribution, scale=10")
plt.show()
```

```python
# Standard Cauchy distribution
cau = pd.Series(np.random.standard_cauchy(100))

plt.hist(cau, color=main_color, bins=50)
plt.axvline(cau.mean(), color=reference_color, linewidth=line_width)
plt.axvline(cau.mean() + cau.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.axvline(cau.mean() - cau.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.show()
```

__Notes on the Cauchy distribution:__ Interesting distribution. The mean and standard deviation seem to be useful estimates of central tendency and deviation in most of the samplings, but the distribution also seems highly volatile, with some of the sampling instances exhibiting right or left skews.

```python
# Plotting the sum of two normal distributions

norm = pd.DataFrame()
norm["small"] = np.random.normal(5, 0.5, 100)
norm["large"] = np.random.normal(10, 1, 100)
norm["comb"] = norm.small + norm.large

plt.hist(norm.comb, color=main_color, bins=20)
plt.axvline(norm.comb.mean(), color=reference_color, linewidth=line_width)
plt.axvline(norm.comb.mean() + norm.comb.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.axvline(norm.comb.mean() - norm.comb.std(), color=reference_color, linewidth=line_width, linestyle="dashed")
plt.title("Adding Normal Distributions with Different Means and Standard Deviations")
plt.show()
```

__Descriptive statistics of the new distribution:__ The mean and standard deviation have the same drawback here as with some of the other distributions: summing these two distributions with different means has resulted in a distribution that isn't truly symmetric about its new mean. That being said, the mean and standard deviation aren't as egregiously inappropriate here as in the case of the F distribution with few degrees of freedom in the numberator and denominator. If I were given a dataset that exhibited a distribution like this I would be wondering if this is normal enough to use linear methods.
